```python
#improt the necessary library necessary for the initial step
import pandas as pd
#using the path of the excel file, call the file
data = pd.read_excel (r'C:\Users\USER\Documents\IBM DataScience\world analysis\world population\work.xlsx', skiprows=range(2)) 
#the first of the data is called
data.head()
```

```python
#what then is the population of the world in the last five years from 2020 descending
ans = data.iloc[1,1], data.iloc[1,2], data.iloc[1,3], data.iloc[1,4], data.iloc[1,5]
ans
```
